```python
train=df.drop(
    [
 #   'year',
   'certificate',
    'movie', 

    'genre', 
  #   'rate',
     'metascore', 
     #'votes',
       'gross', 
     'user_reviews',
     'critic_reviews',
     'popularity', 
     #'awards_wins',
     'awards_nominations',
 'Oscar_Best_Picture_won',
#'Oscar_Best_Picture_nominated', 
    'Oscar_Best_Director_won',
      
   #'Oscar_Best_Director_nominated',
    'Oscar_Best_Actor_won',
    #'Oscar_Best_Actor_nominated', 
    'Oscar_Best_Actress_won',
   #'Oscar_Best_Actress_nominated',
    'other_bestPicture_won',
     #'other_director_won', 
    #'other_actor_won', 
   'other_actress_won',
     'other_bestPicture_nominated', 
     'other_director_nominated',
       'other_actor_nominated',
     'other_actress_nominated'
    ],
     
     
     
     
     
    axis=1)
```

```python
test=pd.DataFrame()
```

```python
test['Oscar_Best_Director_won']=df['Oscar_Best_Director_won']
```

```python
X_train=[]
Y_train=[]
X_test=[]
Y_test=[]
for i in train.index:
    if(int(train['year'][i])<2017):
        print(train.iloc[i])
        X_train.append(train.iloc[i])
        Y_train.append(test['Oscar_Best_Director_won'][i])
    else:
        X_test.append(train.iloc[i])
        Y_test.append(test['Oscar_Best_Director_won'][i])
        
```

```python
X_train=pd.DataFrame(X_train)
Y_train=pd.DataFrame(Y_train)
X_test=pd.DataFrame(X_test)
Y_test=pd.DataFrame(Y_test)
```

```python
X_train=(X_train-X_train.min())/(X_train.max()-X_train.min())
Y_train=(Y_train-Y_train.min())/(Y_train.max()-Y_train.min())
X_test=(X_test-X_test.min())/(X_test.max()-X_test.min())
Y_test=(Y_test-Y_test.min())/(Y_test.max()-Y_test.min())
```

```python
X_train=X_train.drop(['year'],axis=1)
X_test=X_test.drop(['year'],axis=1)
```

```python
regr=LinearRegression()
```

```python
regr.fit(X_train,Y_train)
```

```python
pred=regr.predict(X_test)
```

```python
regr.score(X_test,Y_test)
```

```python
df_norm=pd.DataFrame(pred)
df_norm=(df_norm-df_norm.min())/(df_norm.max()-df_norm.min())
```
