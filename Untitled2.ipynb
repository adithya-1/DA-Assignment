```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('DA Assignment Initial DataSet - Sheet1.csv')
```

```python
picAwardList=[]
picNomList=[]
dirAwardList=[]
dirNomList=[]
actAwardList=[]
actNomList=[]
actressAwardList=[]
actressNomList=[]
```

```python
col=['Golden_Globes_nominated_categories','BAFTA_nominated_categories','Critics_Choice_nominated_categories','Los_Angeles_Film_Critics_Association_nominated_categories','New_York_Film_Critics_Circle_nominated_categories','Boston_Society_of_Film_Critics_nominated_categories','Denver_Film_Critics_Society_nominated_categories','Austin_Film_Critics_Association_nominated_categories','Online_Film_Critics_Society_nominated_categories','London_Critics_Circle_Film_nominated_categories','Hollywood_Film_nominated_categories','Online_Film_Television_Association_nominated_categories']
dire=['Best Director','David Lean Award for Direction','Best Director','Best Director','Best Director','Best Director','Best Director','Best Director','Best Director','Director of the Year','Director of the Year','Best Director']
pic=['Best Motion Picture','Best Film','Best Picture','Best Picture','Best Film','Best Film','Best Picture','Best Film','Best Picture','Film of the Year','Movie of the Year','Best Picture']
act=['Best Performance by an Actor','Best Performance by an Actor','Best Actor','Best Actor','Best Actor','Best Actor','Best Actor','Best Actor','Best Actor','Actor of the Year','Actor of the Year','Best Actor']
actress=['Best Performance by an Actress','Best Performance by an Actress','Best Actress','Best Actress','Best Actress','Best Actress','Best Actress','Best Actress','Best Actress','Actress of the Year','Actress of the Year','Best Actress']

print(len(col),len(dire),len(pic),len(act),len(actress))
```

```python
for co in col:
    df[co].fillna('0',inplace=True)
```

```python
for i in df.index:
    #picAward=0
    picNom=0
    #dirAward=0
    dirNom=0
    #actAward=0
    actNom=0
    #actressAward=0
    actressNom=0
    
    for j in range(len(col)):
        listi=df[col[j]][i].split('|')
    
        if(dire[j] in listi):
            dirNom+= 1+ df['Directors_Guild_nominated'][i]

        if(pic[j] in listi):
            picNom+=1 

        if(act[j] in listi):
            actNom+=1

        if(actress[j] in listi):
            actressNom+=1
    picAwardList.append(picNom)
    #picNomList.append(picNom)
    dirAwardList.append(dirNom)
    #dirNomList.append(dirNom)
    actAwardList.append(actNom)
    #actNomList.append(actNom)
    actressAwardList.append(actressNom)
    #actressNomList.append(actressNom)
    
    
    

    
        
        
    
```

```python
print(picAwardList,
picNomList,
dirAwardList,
dirNomList,
actAwardList,
actNomList,
actressAwardList,
actressNomList)
```

```python
df['other_bestPicture_nominated']=picAwardList
df['other_director_nominated']=dirAwardList
df['other_actor_nominated']=actAwardList
df['other_actress_nominated']=actressAwardList
```

```python
df.to_csv('newProcessed.csv')
```
